# Survival Count

```python
titanic=sns.load_dataset("titanic")
sns.countplot(x='survived', data=titanic)
plt.title("Survival Count")
plt.show()
```

### Insights: Over 500 people dies and over 300 people survived

# Survival Count By Gender

```python
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival by Gender")
plt.show()
```

### Insights: Over 100 males survived and Over 400 males dies
### Over 50 Females survived and Over 200 females died

# Survived By Passenger Class

```python
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title("Survival by Passenger Class")
plt.show()
```

### Insights: passenger class 1:over 50 people died and over 100 survived
### passenger class 2: over 50 people died and over 50 people survied
### passenger class 3: over 100 people survived and over 350 people died

# Age Distribution

```python
plt.figure(figsize=(10, 5))
sns.histplot(titanic['age'].dropna(), kde=True, bins=30)
plt.title("Age Distribution")
plt.show()
```

### Insights: Most People were of age between 20 and 30

# Correlation HeatMap:

```python
plt.figure(figsize=(10, 6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

### Insights: This Heatmap shows correlations of all numeric variables of titanic Dataset:
          1)light red color shows weak positive correlation 
          2)Strong red color shows strong positive correlation
          3)light blue color shows weak negative correlation 
          4)Strong blue color shows strong negative correlation
