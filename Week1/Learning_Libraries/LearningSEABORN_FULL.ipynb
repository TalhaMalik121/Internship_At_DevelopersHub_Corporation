### 1Ô∏è‚É£ Basic Scatter Plot

```python

plt.figure(figsize=(6, 4))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

### 2Ô∏è‚É£ Scatter Plot with Hue

```python

plt.figure(figsize=(6, 4))
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title("Total Bill vs Tip (colored by sex)")
plt.show()
```

### 3Ô∏è‚É£ Histogram (distribution plot)

```python

plt.figure(figsize=(6, 4))
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

### 4Ô∏è‚É£ Box Plot

```python

plt.figure(figsize=(6, 4))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

### 5Ô∏è‚É£ Violin Plot

```python

plt.figure(figsize=(6, 4))
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

### 6Ô∏è‚É£ Bar Plot

```python

plt.figure(figsize=(6, 4))
sns.barplot(x='day', y='tip', data=tips, estimator=sum)
plt.title("Total Tips by Day")
plt.show()
```

### 7Ô∏è‚É£ Count Plot (like a bar plot of counts)

```python

plt.figure(figsize=(6, 4))
sns.countplot(x='day', data=tips)
plt.title("Count of Records per Day")
plt.show()
```

### 8Ô∏è‚É£ Heatmap (correlation matrix)

```python

plt.figure(figsize=(8, 6))
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### 9Ô∏è‚É£ Pairplot (scatter matrix)

```python

sns.pairplot(tips, hue='sex')
plt.suptitle("Pairplot of Tips Dataset", y=1.02)#y=1.02 for tilt the title slightly above the plot for clear visualization
plt.show()
```
