# üìä Learn Matplotlib - Step-by-Step

```python
import matplotlib.pyplot as plt
import numpy as np

# Display plots inline (only needed in Jupyter)
%matplotlib inline
```

# 1Ô∏è‚É£ Basic Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))#create a figure 8 inch wide and 4 inch tall
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', linewidth=2)#ploy y vs x with blue color solid thicker line
plt.title("Basic Sine Wave Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()#add a legend box 
plt.grid(True)#add grid bg for better readability
plt.show()
```

# 2Ô∏è‚É£ Multiple Lines on One Plot

```python
y2 = np.cos(x)
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='red')
plt.plot(x, y2, label='cos(x)', color='green')
plt.title("Sin and Cos Waves")
plt.legend()
plt.grid(True)
plt.show()
```

# 3Ô∏è‚É£ Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(6, 6))
plt.scatter(x, y, color='purple', marker='o')
plt.title("Random Scatter Plot")
plt.xlabel("Random X")
plt.ylabel("Random Y")
plt.show()
```

# 4Ô∏è‚É£ Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='orange')
plt.title("Simple Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

# 5Ô∏è‚É£ Histogram

```python

data = np.random.randn(1000)

plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# 6Ô∏è‚É£ Pie Chart

```python

sizes = [25, 35, 20, 20]
labels = ['Python', 'Java', 'C++', 'JavaScript']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Popularity")
plt.axis('equal')  # Equal aspect ratio ensures a circle.
plt.show()
```
